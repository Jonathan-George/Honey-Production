```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\JONATHAN GEORGE\Downloads\honeyproduction.csv")
```

```python
print(df.head())
```

```python
prod_per_year = df.groupby('year').totalprod.mean().reset_index()
```

```python
x = df['year']
```

```python
x = x.values.reshape(-1, 1)
```

```python
y = df['totalprod']
```

```python
y = y.values.reshape(-1, 1)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = linear_model.LinearRegression()
```

```python
regr = LinearRegression()
regr.fit(x, y)
```

```python
# Print out the slope of the line (coefficients)
print("Slope of the line (Coefficients):", regr.coef_)
```

```python
# Print out the intercept of the line
print("Intercept of the line:", regr.intercept_)
```

```python
y_predict = regr.predict(x)
```

```python
plt.plot(x, y_predict)

plt.show()
```
